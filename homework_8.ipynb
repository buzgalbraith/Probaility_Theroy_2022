-our goal is to find $f_{a,b}(a,b)$ we are going to do that as $f_{a,b}(a,b)=f_{b}(b)f_{a|b}(a|b)$ 
- first off we know that our traingle has area $\frac{1}{2}(1)(1)=\frac{1}{2}$ thus we can see that $\int_{0}^{.5}\int_{0}^{2x}cdydx+\int_{.5}^{1}\int_{0}^{2-2x}cdydx=\frac{1}{4}c+\frac{1}{4}c=\frac{1}{2}c$ then we want that to integrate to 1. so c=2
- in other words we know that our joint cdf takes value 2 everywhere in the triangle. 
- now we want to think about $f_{x,y}(x,y)=f_x(x)f_{y|x}(y|x)$
- item so we can see that $f_{y|x}(y|x)=\frac{f_{y,x}(y,x)}{f_{x}(x)}$
- now we want to marginalize over one axis. $f(x)=\int_{y=0}^{y=2x}f_{x,y}(x,y)dy+\int_{y=0)}^{y=2-2x}f_{x,y}(x,y)dy=\int_{y=0}^{y=2x}\frac{1}{5}dy+\int_{y=0)}^{y=2-2x}\frac{1}{5}dy=\frac{2}{5}$  
- ok now we want to take the conditional $f(y|x)=\frac{f(x,y)}{f(x)}$=$\frac{2}{}$

here we are genearing two random samples of length 1000 and plotting them. 

```python
u_1,u_2=np.random.rand(2,1000)
plt.scatter(u_1,u_2)
```

```python
def inverse_x(u_1):
    out=[]
    for x in u_1:
        if(x<.5):
            out.append(np.sqrt(x/2))
        else:
            out.append(1-np.sqrt(8*(1-x))/4)
    return out
def inverse_y(u_2,x):
    out=[]
    for i in range(len(x)):
        if(x[i]<.5):
            out.append(u_2[i]*(2*x[i]))
        else:
            out.append(u_2[i]*((2-2*x[i])))
    return out
x=inverse_x(u_1)
y=inverse_y(u_2,x)
plt.scatter(x,y)
```
